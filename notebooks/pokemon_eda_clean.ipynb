## ‚öîÔ∏è 3. Type Matchups & Effectiveness

But what if the Types that a Pokemon holds should also affect our ranking? Are there types which perform better than others, because they are more so often effective or resistant against other types? And how can we add this to our ranking? 

First we need to store this information regarding type matchups and effectiveness in order to calculate whether certain types perform better than others.


```python
# Simplified Gen 1 type chart (values: 2 = super effective, 0.5 = not very effective, 0 = no effect)
type_chart = {
    'Normal':     {'Rock': 0.5, 'Ghost': 0},
    'Fire':       {'Fire': 0.5, 'Water': 0.5, 'Grass': 2, 'Ice': 2, 'Bug': 2, 'Rock': 0.5, 'Dragon': 0.5},
    'Water':      {'Fire': 2, 'Water': 0.5, 'Grass': 0.5, 'Ground': 2, 'Rock': 2, 'Dragon': 0.5},
    'Electric':   {'Water': 2, 'Electric': 0.5, 'Grass': 0.5, 'Ground': 0, 'Flying': 2, 'Dragon': 0.5},
    'Grass':      {'Fire': 0.5, 'Water': 2, 'Grass': 0.5, 'Poison': 0.5, 'Ground': 2, 'Flying': 0.5, 'Bug': 0.5, 'Rock': 2, 'Dragon': 0.5},
    'Ice':        {'Fire': 0.5, 'Water': 0.5, 'Grass': 2, 'Ground': 2, 'Flying': 2, 'Dragon': 2},
    'Fighting':   {'Normal': 2, 'Ice': 2, 'Rock': 2, 'Ghost': 0, 'Poison': 0.5, 'Flying': 0.5, 'Psychic': 0.5, 'Bug': 0.5},
    'Poison':     {'Grass': 2, 'Poison': 0.5, 'Ground': 0.5, 'Bug': 2, 'Rock': 0.5, 'Ghost': 0},
    'Ground':     {'Fire': 2, 'Electric': 2, 'Grass': 0.5, 'Poison': 2, 'Flying': 0, 'Bug': 0.5, 'Rock': 2},
    'Flying':     {'Grass': 2, 'Fighting': 2, 'Bug': 2, 'Electric': 0.5, 'Rock': 0.5},
    'Psychic':    {'Fighting': 2, 'Poison': 2, 'Psychic': 0.5},
    'Bug':        {'Fire': 0.5, 'Grass': 2, 'Fighting': 0.5, 'Poison': 0.5, 'Flying': 0.5, 'Psychic': 2, 'Ghost': 0.5},
    'Rock':       {'Fire': 2, 'Ice': 2, 'Flying': 2, 'Bug': 2, 'Fighting': 0.5, 'Ground': 0.5},
    'Ghost':      {'Normal': 0, 'Psychic': 2, 'Ghost': 2},
    'Dragon':     {'Dragon': 2},
}
```

Here is a visualization of the matchup structure. The types along the y-axis represent the attacking type, while the types along the x-axis represent the defending type. This means that the rows show how effective a type is versus another. The columns show how vulnerable a type is versus another (0.0 being complete immunity).

```python
# Full list of Gen 1 types
gen1_types = [
    'Normal', 'Fire', 'Water', 'Electric', 'Grass', 'Ice',
    'Fighting', 'Poison', 'Ground', 'Flying', 'Psychic',
    'Bug', 'Rock', 'Ghost', 'Dragon'
]

# Initialize empty matrix with neutral values (1.0)
type_matrix = pd.DataFrame(
    1.0, index=gen1_types, columns=gen1_types
)

# Populate with effectiveness values
for attacker, defenders in type_chart.items():
    for defender, multiplier in defenders.items():
        type_matrix.loc[attacker, defender] = multiplier

# Set up custom colormap for better contrast
cmap = sns.color_palette("vlag", as_cmap=True)

# Plot the heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(
    type_matrix,
    annot=True,
    fmt=".1f",
    cmap=cmap,
    linewidths=0.5,
    linecolor='gray',
    cbar_kws={"label": "Effectiveness Multiplier"},
    square=True
)

plt.title("Gen 1 Type Matchup Matrix", fontsize=16, weight='bold')
plt.xlabel("Defending Type", fontsize=12)
plt.ylabel("Attacking Type", fontsize=12)
plt.xticks(rotation=45)
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```

## üßÆ 4. Calculating Type Scores
Next, we calculate the Type Score by paying attention to both how a type performs while attacking certain other types, but also while defending against other types.

```python
# List of all Gen 1 types
all_types = list(type_chart.keys())

def offensive_score(pokemon_types):
    """Average effectiveness of a type against all others"""
    score = 0
    for atk_type in pokemon_types:
        for def_type in all_types:
            multiplier = type_chart.get(atk_type, {}).get(def_type, 1.0)
            score += multiplier
    return score / (len(pokemon_types) * len(all_types))

def defensive_score(pokemon_types):
    """Average resistance/vulnerability (1 / incoming multipliers from all types)"""
    score = 0
    for def_type in pokemon_types:
        for atk_type in all_types:
            multiplier = type_chart.get(atk_type, {}).get(def_type, 1.0)
            score += 1 / (multiplier if multiplier != 0 else 0.2)  # treat immunity as stronger than resistance (0.5)
    return score / (len(pokemon_types) * len(all_types))
```

```python
def get_type_score(row):
    types = [row['Type 1']]
    if pd.notna(row['Type 2']):
        types.append(row['Type 2'])
    
    off = offensive_score(types)
    defn = defensive_score(types)
    
    total_type_score = off + defn
    return round(total_type_score * 50, 2)  # Scale up for contribution to Total
 
df['TypeScore'] = df.apply(get_type_score, axis=1)
df['AdjustedTotal'] = df['Total'] + df['TypeScore']
```

```python
# Compute score for each individual type (assume mono-type)
type_scores = []

for t in all_types:
    off = offensive_score([t])
    defn = defensive_score([t])

    score = off + defn
    type_scores.append({
        'Type': t,
        'OffensiveScore': round(off, 3),
        'DefensiveScore': round(defn, 3),
        'TotalTypeScore': round(score * 50, 2)  # scale for comparability
    })

type_df = pd.DataFrame(type_scores).sort_values(by='TotalTypeScore', ascending=False)
```

```python
# Function to build type icon URL
def get_type_icon_url(type_name):
  return f"https://play.pokemonshowdown.com/sprites/types/{type_name}.png"

# Function to build a single row with all relevant visuals and stats
def build_row_html(row):
    type1_icon = f'<img src="{get_type_icon_url(row["Type"])}" width="50">'
    types_html = f"{type1_icon}"

    return pd.Series({
        'Types': types_html,
        'OffensiveIndex' : row['OffensiveScore'],
        'DefensiveIndex' : row['DefensiveScore'],
        'TotalTypeScore' : row['TotalTypeScore']
    })

# Apply to first 9 Pok√©mon
visual_table = type_df.apply(build_row_html, axis=1)

# Show the visual table as HTML
HTML(visual_table.to_html(escape=False, index=False))
```

Types matter. And certain types performs visible better than other. Most dominantly types that have a good defense against other types such as Ghost or Flying, since they are completely immune to certain other attacking types. With this we can visualize the matchup structure again, but with the rows and columns now sorted: the strongest attackers are at the top and the strongest defenders to  the left.

```python
# Step 1: Create the type matchup DataFrame
all_types = sorted(set(type_chart.keys()).union(*[set(v.keys()) for v in type_chart.values()]))
df_chart = pd.DataFrame(1.0, index=all_types, columns=all_types)

for attacker, defenders in type_chart.items():
    for defender, value in defenders.items():
        df_chart.loc[attacker, defender] = value

# Step 2: Calculate effectiveness scores
attack_strength = df_chart.mean(axis=1)  # Row-wise: Attacking effectiveness
defense_weakness = df_chart.mean(axis=0)  # Column-wise: Defensive vulnerability

# Step 3: Sort rows (attackers) and columns (defenders)
sorted_attack = attack_strength.sort_values(ascending=False).index
sorted_defense = defense_weakness.sort_values(ascending=True).index
df_sorted = df_chart.loc[sorted_attack, sorted_defense]

# Step 4: Plot the heatmap
plt.figure(figsize=(12, 10))
sns.set(style="whitegrid")

ax = sns.heatmap(
    df_sorted,
    cmap=cmap,
    annot=True,
    fmt=".1f",
    linewidths=0.5,
    square=True,
    cbar_kws={'label': 'Effectiveness Multiplier'},
    linecolor='gray'
)

plt.title("Gen 1 Type Matchup Matrix (Sorted by Avg Effectiveness)", fontsize=16, weight='bold', pad=15)
plt.xlabel("Defending Type (least vulnerable ‚Üí most vulnerable)", fontsize=12)
plt.ylabel("Attacking Type (least effective ‚Üí most effective)", fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)

plt.tight_layout()
plt.show()
```

> ‚ö†Ô∏è Note on Defensive Scores:
> You may notice that some types like **Flying** perform better in the prior ranking table than others like **Dragon**, even though the heatmap here may suggest otherwise.
>
> This is because the score table **rewards immunities and resistances** more heavily by using an inverse multiplier system (`1 / effectiveness`, with immunities counted as `1 / 0.2 = 5`), while the heatmap uses the **raw type chart values**.
>
> These are two valid but distinct ways of looking at type strength ‚Äî one emphasizes raw interactions, the other strategic advantage.


## üß¨ 5. Adjusting Total Stats with Type Score
Next we want to simply add the TotalTypeScore to the Total Score of every Pokemon. We already calculated the Total Type Score so that it can have some impact when added to the Total Score.

However we have one problem: pokemon sometimes have 2 types. Our model needs to account for that. Hence we adjust it so that attacking Pokemon can choose the most effective of its two types, and defending types multiply: e.g. if both defending types are resistant against the attacking type, the damage gets cut in half two times.

```python
def offensive_score_dual(types):
    score = 0
    for def_type in all_types:
        effectiveness = []
        for atk_type in types:
            effectiveness.append(type_chart.get(atk_type, {}).get(def_type, 1))
        score += max(effectiveness)  # Take best offensive coverage
    return score / len(all_types)
```

```python
def defensive_score_dual(types):
    score = 0
    for atk_type in all_types:
        mult = 1
        for def_type in types:
            mult *= type_chart.get(atk_type, {}).get(def_type, 1)
        #mult = max(mult, 0.2) #or clamp here
        score += 1 / (mult if mult > 0 else 0.2)  # Reward immunities more than double resistance
    return score / len(all_types)
```

```python
def get_type_score_dual(row):
    types = [row['Type 1']]
    if pd.notna(row['Type 2']):
        types.append(row['Type 2'])

    off = offensive_score_dual(types)
    defn = defensive_score_dual(types)

    return round((off + defn) * 50, 2)  # scale to match Total stat
```

```python
df['TypeScore'] = df.apply(get_type_score_dual, axis=1)
df['AdjustedTotal'] = df['Total'] + df['TypeScore']
```

```python
def get_sprite_url(poke_id):
    return f'https://raw.githubusercontent.com/PokeAPI/sprites/master/sprites/pokemon/{poke_id}.png'

def get_type_icon_url(type_name):
    return f"https://play.pokemonshowdown.com/sprites/types/{type_name}.png"

def build_row_html(row):
    sprite = f'<img src="{get_sprite_url(row.name)}" width="50">'
    type1_icon = f'<img src="{get_type_icon_url(row["Type 1"])}" width="40">'
    type2_icon = f'<img src="{get_type_icon_url(row["Type 2"])}" width="40">' if pd.notna(row["Type 2"]) else ""
    types_html = f"{type1_icon}{type2_icon}"
    return pd.Series({
        'Sprite': sprite,
        'Name': row['Name'],
        'Types': types_html,
        'TotalScore': row['Total']
    })

def build_row_html_adj(row):
    sprite = f'<img src="{get_sprite_url(row.name)}" width="50">'
    type1_icon = f'<img src="{get_type_icon_url(row["Type 1"])}" width="40">'
    type2_icon = f'<img src="{get_type_icon_url(row["Type 2"])}" width="40">' if pd.notna(row["Type 2"]) else ""
    types_html = f"{type1_icon}{type2_icon}"
    return pd.Series({
        'Sprite': sprite,
        'Name': row['Name'],
        'Types': types_html,
        'AdjustedTotalScore': row['AdjustedTotal']
    })

# Original top 10
top10_all = df[(df['Legendary'] == False) & (df['Name'] != 'Mew')].sort_values(by='Total', ascending=False).head(10).apply(build_row_html, axis=1)

# Top 10 Adjusted
filtered_df = df[(df['Legendary'] == False) & (df['Name'] != 'Mew')]
top10_filtered = filtered_df.sort_values(by='AdjustedTotal', ascending=False).head(10).apply(build_row_html_adj, axis=1)

# Convert to HTML
html_all = top10_all.to_html(escape=False, index=False)
html_filtered = top10_filtered.to_html(escape=False, index=False)

# Combine the two tables side by side
combined_html = f"""
<div style="display: flex; gap: 40px;">
    <div>
        <h3>Top 10 Gen 1 Pok√©mon (Non-Legendaries)</h3>
        {html_all}
    </div>
    <div>
        <h3>Top 10 Gen 1 Pok√©mon Adjusted (Non-Legendaries)</h3>
        {html_filtered}
    </div>
</div>
"""

HTML(combined_html)
```

We can now compare our ranking list where on the left we have the order accoring to the original Total Score of the summed six stats, and on the right the Adjusted Total Score where we added the Total Type Score. Indeed, Dragonite remains at the top. But we have now some new Pokemon among the Top 10 that hold very strong Types like Ghost, and Pokemon that have Flying among their Types get ranked now higher than before. 


Next we might ask: what are generally the strongest combination of types and which Pokemon represent them?

```python
import itertools

# Step 1: Generate all mono and dual combinations
mono_types = [(t,) for t in all_types]
dual_types = list(itertools.combinations(all_types, 2))
all_combinations = mono_types + dual_types

# Step 2: Define a helper function to calculate scores for any type combo
def calculate_dual_type_score(types):
    off = offensive_score_dual(types)
    defn = defensive_score_dual(types)
    return round((off + defn) * 50, 2)

# Step 3: Compute scores for all combos
combination_scores = []
for combo in all_combinations:
    score = calculate_dual_type_score(list(combo))
    combination_scores.append({
        'TypeCombo': '/'.join(combo),
        'TotalTypeScore': score
    })

# Step 4: Create DataFrame and sort
combo_df = pd.DataFrame(combination_scores).sort_values(by='TotalTypeScore', ascending=False)
```

```python
# Step 1: Create a 'TypeCombo' column in your df that sorts and joins the types
def get_type_combo(row):
    types = [row['Type 1']]
    if pd.notna(row['Type 2']):
        types.append(row['Type 2'])
    return '/'.join(sorted(types))

df['TypeCombo'] = df.apply(get_type_combo, axis=1)

# Step 2: For each TypeCombo, find the Pok√©mon with the highest Total stat
best_pokemon_per_combo = df.loc[df.groupby('TypeCombo')['Total'].idxmax()][['TypeCombo', 'Name', 'Total']]

# Step 3: Calculate scores for existing combos as before
existing_combos = best_pokemon_per_combo['TypeCombo'].tolist()

existing_combo_scores = []
for combo in existing_combos:
    types = combo.split('/')
    score = calculate_dual_type_score(types)
    existing_combo_scores.append({
        'TypeCombo': combo,
        'TotalTypeScore': score
    })

existing_combo_df = pd.DataFrame(existing_combo_scores)

# Step 4: Merge scores with best Pok√©mon info
result_df = existing_combo_df.merge(best_pokemon_per_combo, on='TypeCombo')

# Step 5: Sort by TotalTypeScore descending
result_df = result_df.sort_values(by='TotalTypeScore', ascending=False).reset_index(drop=True)
```

```python
# Sprite fetch function
def get_sprite_url(dex_id):
    return f'https://raw.githubusercontent.com/PokeAPI/sprites/master/sprites/pokemon/{dex_id}.png'

# Type icon fetch function
def get_type_icon_url(type_name):
    return f'https://play.pokemonshowdown.com/sprites/types/{type_name}.png'

# Name ‚Üí Dex ID mapping
name_to_id = df['Name'].to_dict()
name_to_dex = {v: k for k, v in name_to_id.items()}

# Name ‚Üí Type1 and Type2 mapping
type_lookup = df.set_index('Name')[['Type 1', 'Type 2']].to_dict('index')

# Add sprite and build HTML row
def build_combo_row(row):
    dex_id = name_to_dex.get(row['Name'], 1)  # fallback to 1
    sprite = f'<img src="{get_sprite_url(dex_id)}" width="50">'
    
    # Extract types from lookup and build icons
    type1 = type_lookup[row['Name']]['Type 1']
    type2 = type_lookup[row['Name']]['Type 2']
    type_icons = f'<img src="{get_type_icon_url(type1)}" width="40">'
    if pd.notna(type2):
        type_icons += f'<img src="{get_type_icon_url(type2)}" width="40">'
    
    return pd.Series({
        'TypeCombo': type_icons,
        'TotalTypeScore': row['TotalTypeScore'],
        'Total': row['Total'],
        'Best Representative': row['Name'],
        'Sprite': sprite
    })

# Apply to top 10
visual_result_df = result_df.head(10).apply(build_combo_row, axis=1)

# Display as HTML
HTML(visual_result_df.to_html(escape=False, index=False))
```

At the top we see Ghost/Prison with Gengar. Rock/Ground and Poison/Flying are strong as well, but Golem and Golbat do not have a high Total Score to be seen among the upper Top 10. 

Now what about combinations that did not yet exist in Gen 1? Say we want to design more Pokemon in the future; with which combination of types should we be more careful?

```python
# Filter out combos containing 'Ghost'
all_combinations_no_ghost = [combo for combo in all_combinations if 'Ghost' not in combo]

# Then compute scores as before, but on filtered combos
combination_scores_no_ghost = []
for combo in all_combinations_no_ghost:
    score = calculate_dual_type_score(list(combo))
    combination_scores_no_ghost.append({
        'TypeCombo': '/'.join(combo),
        'TotalTypeScore': score
    })

combo_df_no_ghost = pd.DataFrame(combination_scores_no_ghost).sort_values(by='TotalTypeScore', ascending=False)
```

```python
# Function to convert a type string into an <img> tag
def type_to_icon(type_name):
    return f'<img src="https://play.pokemonshowdown.com/sprites/types/{type_name}.png" width="40" style="margin-right:4px;">'

# Function to convert a 'TypeCombo' string like 'Fire/Flying' into icons
def type_combo_to_icons(combo):
    types = combo.split('/')
    return ''.join([type_to_icon(t) for t in types])

# Make copies to avoid modifying original
combo_df_vis = combo_df.head(10).copy()
combo_df_no_ghost_vis = combo_df_no_ghost.head(10).copy()

# Replace TypeCombo with icons
combo_df_vis['TypeCombo'] = combo_df_vis['TypeCombo'].apply(type_combo_to_icons)
combo_df_no_ghost_vis['TypeCombo'] = combo_df_no_ghost_vis['TypeCombo'].apply(type_combo_to_icons)

# Convert to HTML with escaping disabled to allow image rendering
left_table_html = combo_df_vis.to_html(escape=False, index=False)
right_table_html = combo_df_no_ghost_vis.to_html(escape=False, index=False)

# Combine into a side-by-side layout
html = f"""
<div style="display: flex; justify-content: space-around; gap: 20px;">
  <div>
    <h3>Best Type Combinations</h3>
    {left_table_html}
  </div>
  <div>
    <h3>Best Type Combinations (No Ghost)</h3>
    {right_table_html}
  </div>
</div>
"""

# Display in Jupyter
display(HTML(html))
```

## üå° 6. Correlation Analysis
Since we already talked about designing new Pokemon, it is good to take a look at the open design space. We don't want our novel Pokemon to be too similar to the ones that already exist. 

For this we take another look at the stats individually. Which stats have a strong or weak correlation among the existing 151 Pokemon? E.g. A high value in stat 1 implies a high value for stat 2. 

```python
# Compute correlation matrix
correlation_matrix = df_stats.iloc[:, 3:].corr()

# Upper triangle only
upper_triangular = correlation_matrix.where(np.triu(np.ones(correlation_matrix.shape), k=1).astype(bool))
correlation_pairs = upper_triangular.stack().sort_values(ascending=False)

# Create the figure
fig, axes = plt.subplots(1, 2, figsize=(18, 8), gridspec_kw={'width_ratios': [2, 1]})

# --- Left: Heatmap ---
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap="afmhot", cbar=True, ax=axes[0])
axes[0].set_title("Correlation Matrix")

# --- Right: Tabular Text Display ---
axes[1].axis("off")

top_n = 15
rows = [f"{'Rank':<5} {'Var 1':<12} {'Var 2':<12} {'Corr':>6}"]
rows.append("-" * 38)

for i, ((col1, col2), val) in enumerate(correlation_pairs.head(top_n).items(), start=1):
    rows.append(f"{i:<5} {col1:<12} {col2:<12} {val:>6.2f}")

tabular_text = "\n".join(rows)
axes[1].text(0, 1, f"Top {top_n} Correlations\n\n{tabular_text}", fontsize=12, va='top', ha='left', family='monospace')

plt.show()
```

It is apparent that Attack and Special Attack always correlate with their defensive counterparts. This means, if a Pokemom has a high Attack, it will most likely have a high Defense as well. More interesting are pairs with low correlation. Here we find possible future design space. HP and Defense have a very weak correlation with Speed. Hence, we can learn from this, that it might be interesting to design future Pokemon that hold a high HP/Defense in addition to being fast.

**A few considerations emerged for further discussion or refinement:**

- ‚öñÔ∏è **Is immunity too powerful in the Type Score?**  
  The current scoring grants significant weight to type immunities. If this proves to skew rankings, the scoring system could be **rebalanced** by reducing the points assigned to immunities.

- üìà **Should type effectiveness consider encounter frequency?**  
  Currently, type matchups are treated uniformly. But in practice, some types are far more common than others. Adjusting scores based on **how frequently types appear** in the dataset could better reflect real-world effectiveness.

- üß¨ **What if we account for dominance dynamics?**  
  Being strong or resistant against a particularly dominant type might deserve more weight. This would make the scoring **more contextual**, rewarding counters to overpowered or popular types.

- üß† **Insights for future Pok√©mon design:**  
  This analysis highlighted areas of **open design space** ‚Äî underused combinations of stat distributions that new Pok√©mon could explore. It also brought to light recurring **design philosophies**, helping future creations remain faithful to the legacy of the franchise.
